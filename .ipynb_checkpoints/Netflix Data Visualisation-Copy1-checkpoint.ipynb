```python
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# Define the base directory and file name
base_dir = Path.home()  # This points to the user's home directory
#print (base_dir) #Error checking code
file_name = 'netflix_data.csv'

path= base_dir / 'Downloads' / file_name

print (path) #Error checking code

#Define function to check if file path exists
def is_valid_file_path(file_path):
    # Check if the path exists and is a file
    if os.path.exists(file_path) and os.path.isfile(file_path):
        return True
    else:
        return False

if is_valid_file_path(path):
    print("This is a valid file path.")
else:
    print("This is an valid file path.")
```

```python
Netflix_shows_movies= pd.read_csv(path)

# Handling missing values- preprocessing
#Display the number of missing values per column
missing_values = Netflix_shows_movies.isnull().sum()

#Print number of missing values
print(missing_values)

# Display the percentage of missing values per column
missing_percentage = Netflix_shows_movies.isnull().mean() * 100
#print(missing_percentage)

num_columns= len(Netflix_shows_movies.columns)
print("Number of rows:",len(Netflix_shows_movies))
print("Number of columns:",num_columns)

#Replacing missing values for directors, cast, country, date_added and rating
Netflix_shows_movies['director'] = Netflix_shows_movies['director'].fillna('unknown')
Netflix_shows_movies['cast'] = Netflix_shows_movies['cast'].fillna('unknown')
Netflix_shows_movies['country'] = Netflix_shows_movies['country'].fillna('unknown')
Netflix_shows_movies['rating'] = Netflix_shows_movies['rating'].fillna('unknown')
Netflix_shows_movies['date_added'] = Netflix_shows_movies['date_added'].fillna('unknown')

```

```python
#Basic data exploration
#Display first 5 rows
print(Netflix_shows_movies.head())

#Display last 5 rows
print(Netflix_shows_movies.tail())

# Get the number of rows and columns
print(f"Number of rows and columns: {Netflix_shows_movies.shape}")

# List column names
print("Column names:", Netflix_shows_movies.columns)

# List data types of each column
print("\nData types:\n", Netflix_shows_movies.dtypes)

# Summary information about the DataFrame
print(Netflix_shows_movies.info())


#Include some descriptive statistics from the dataframe
# Descriptive statistics for all columns
print(Netflix_shows_movies.describe(include='all'))

```

```python
# Descriptive statistics for numerical columns
print(Netflix_shows_movies.describe())
```

```python
# Check for duplicate show_ids
duplicate_ids = Netflix_shows_movies['show_id'].duplicated().sum()
print(f"Number of duplicate show IDs: {duplicate_ids}")

# Count of total number of shows/movies
total_shows = Netflix_shows_movies['show_id'].nunique()
print(f"Total number of unique shows/movies: {total_shows}")
```

```python
# Calculate the most ocurring netflix content type

mode_value = Netflix_shows_movies['type'].mode()[0]

print(f"Most occurring Netflix content type: {mode_value}")
```

```python
# Count the number of movies and TV shows
type_counts = Netflix_shows_movies['type'].value_counts()
print("Number of Movies vs. TV Shows:\n", type_counts)

# Plot the distribution of types
type_counts.plot(kind='bar', title='Number of Movies vs. TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# Check for duplicate titles
duplicate_titles = Netflix_shows_movies['title'].duplicated().sum()
print(f"Number of duplicate titles: {duplicate_titles}")

# Word frequency analysis
from collections import Counter

title_words = ' '.join(Netflix_shows_movies['title']).split()
word_freq = Counter(title_words)
print("Most common words in titles:", word_freq.most_common(10))

# Length of titles
title_length = Netflix_shows_movies['title'].apply(len)
print(f"Average title length: {title_length.mean()}")
```

```python
# Directors with the most number of movies to their names!!
top_directors = Netflix_shows_movies['director'].value_counts().head(10)
print("Top 10 Most Prolific Directors:\n", top_directors)

# Number of unique directors
unique_directors = Netflix_shows_movies['director'].nunique()
print(f"Number of unique directors: {unique_directors}")
```

```python
# To understand the most common actors on Netflix, we begin by splitting cast into individual names and then determining the frequency
#of appearances in shows/movies
cast_list = Netflix_shows_movies['cast'].dropna().str.split(', ').sum()
cast_freq = Counter(cast_list)
print("Most common actors/actresses:\n", cast_freq.most_common(10))  #Prints the 10 most common actors/actresses
```

```python
# Lets do some analysis based on dates_added
# First convert date_added to datetime
Netflix_shows_movies['date_added'] = pd.to_datetime(Netflix_shows_movies['date_added'], errors='coerce')

# Count the number of shows/movies added by year
added_by_year = Netflix_shows_movies['date_added'].dt.year.value_counts().sort_index()
print("Shows/Movies added by year:\n", added_by_year)

# Plot trends over time
added_by_year.plot(kind='line', title='Content Added Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Shows/Movies Added')
plt.show()
```

```python
# Comparison between release year and date added to find older content being added
#Distribution of shows/movies by release year
release_year_counts = Netflix_shows_movies['release_year'].value_counts().sort_index()
print("Shows/Movies by release year:\n", release_year_counts)

# Plot the release year distribution
release_year_counts.plot(kind='line', title='Content Release Year Distribution')
plt.xlabel('Year')
plt.ylabel('Number of Shows/Movies')
plt.show()
```

```python
# Most watched genres (As per requirement in assignment)
genres_list = Netflix_shows_movies['listed_in'].str.split(', ').sum()
genre_counts = Counter(genres_list)
print("Most watched genres:\n", genre_counts.most_common(10))

# Plot the 10 most watched genres
pd.Series(genre_counts).nlargest(10).plot(kind='bar', title='Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of content ratings (As per requirement in assignment)
rating_counts = Netflix_shows_movies['rating'].value_counts()
print("Distribution of content ratings:\n", rating_counts)

# Plot the distribution of content ratings
rating_counts.plot(kind='bar', title='Content Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
