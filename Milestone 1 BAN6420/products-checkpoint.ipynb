```python
class Product: # A class for insurance policy products
    def __init__(self, product_id, name, description, price):
        try:
            if not product_id or not isinstance(product_id, int):
                raise ValueError("Invalid product ID")
            if not name or not isinstance(name, str):
                raise ValueError("Invalid name")
            if not description or not isinstance(description, str):
                raise ValueError("Invalid description")
            if not price or not isinstance(price, (int, float)):
                raise ValueError("Invalid price")

            self.product_id = product_id
            self.name = name
            self.description = description
            self.price = price
            self.active = True  # Adding an attribute to track if the product is active
        except ValueError as e:
            print(f"Error initializing Product: {e}")
        except Exception as e:
            print(f"Unexpected error: {e}")
        finally:
            print("Initialization attempt completed.")

    def update_price(self, new_price): #Methode to update price of policy 
        try:
            if not new_price or not isinstance(new_price, (int, float)):
                raise ValueError("Invalid price")
            self.price = new_price
        except ValueError as e:
            print(f"Error updating price: {e}")
        except Exception as e:
            print(f"Unexpected error: {e}")
        finally:
            print("Price update attempt completed.")

    def update_description(self, new_description):  #Method to update policy description
        try:
            if not new_description or not isinstance(new_description, str):
                raise ValueError("Invalid description")
            self.description = new_description
        except ValueError as e:
            print(f"Error updating description: {e}")
        except Exception as e:
            print(f"Unexpected error: {e}")
        finally:
            print("Description update attempt completed.")

    def suspend_product(self): #A method to suspend a policy product
        try:
            self.active = False
        except Exception as e:
            print(f"Unexpected error: {e}")
        finally:
            print("Product suspension attempt completed.")

    def activate_product(self): #A method to activate policy product
        try:
            self.active = True
        except Exception as e:
            print(f"Unexpected error: {e}")
        finally:
            print("Product activation attempt completed.")

    def __str__(self):
        status = "Active" if self.active else "Suspended"
        return f"Product({self.product_id}, {self.name}, {self.description}, {self.price}, {status})"

```
