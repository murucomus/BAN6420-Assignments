```python
class Person: #A super class for holding information about a person
    def __init__(self, name, address, phone_number): #A constructor method with the initialisation attributes for each "person' object
        try:
            self.name = name
            self.address = address
            self.phone_number = phone_number
        except Exception as e:
            print(f"Error initializing Person: {e}")
        finally:
            print("Initialization of Person started.")

    def update_address(self, new_address): #method to update addresses
        try:
            self.address = new_address
        except Exception as e:
            print(f"Error updating address: {e}")
        finally:
            print("Attempted to update address.")

    def update_phone_number(self, new_phone_number): #A method to update phone numbers
        try:
            self.phone_number = new_phone_number
        except Exception as e:
            print(f"Error updating phone number: {e}")
        finally:
            print("Attempted to update phone number.")

class Policyholder(Person): #A class for a policy holder but inheriting some attributes from the Person class
    def __init__(self, policyholder_id, name, address, phone_number):
        try:
            super().__init__(name, address, phone_number) #This is to ensure that the Person superclass is properly initialised
            self.policyholder_id = policyholder_id
            self.active = True
        except Exception as e:
            print(f"Error initializing Policyholder: {e}")
        finally:
            print("Initialization of Policyholder started.")

    def register(self): #A method to register new members/policyholders
        try:
            self.active = True
            print(f"Policyholder {self.name} has been registered.")
        except Exception as e:
            print(f"Error registering policyholder: {e}")
        finally:
            print("Attempted to register policyholder.")

    def suspend(self):# A method to suspend existing policyholders
        try:
            self.active = False
            print(f"Policyholder {self.name} has been suspended.")
        except Exception as e:
            print(f"Error suspending policyholder: {e}")
        finally:
            print("Attempted to suspend policyholder.")

    def reactivate(self): #A method to reactivate previously deactivates policyholder accounts
        try:
            self.active = True
            print(f"Policyholder {self.name} has been reactivated.")
        except Exception as e:
            print(f"Error reactivating policyholder: {e}")
        finally:
            print("Attempted to reactivate policyholder.")

    #def __str__(self): #to display policyholder details
       # try:
           # status = "Active" if self.active else "Suspended"
           # return f"Policyholder({self.policyholder_id}, {self.name}, {self.address}, {self.phone_number}, {status})"
        #except Exception as e:
          #  return f"Error converting Policyholder to string: {e}"
        #finally:
           # print("Attempted to convert Policyholder to string.")
    def display_info(self):
        try:
            status = "Active" if self.active else "Suspended"
            return f"Policy Holder Account Details:{self.policyholder_id}, {self.name}, {self.address}, {self.phone_number}, {status}"
        except Exception as e:
            return f"Error converting Policyholder to string: {e}"
        finally:
            print("Attempted to convert Policyholder to string.")
```
